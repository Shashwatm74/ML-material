```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Location1.csv')
df.head()
```

```python
df.drop('Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr())
```

```python
df.isnull().sum()
```

```python
for k in numeric_df:
    if k != 'Power':
        df.groupby([k])['Power'].mean().plot.bar()
        plt.yscale('symlog')  # Change the scale of the y-axis to logarithmic
        plt.xscale('symlog')  # Change the scale of the x-axis to logarithmic
        plt.show()
```

```python
num = df.select_dtypes(exclude = ['object'])
for k in num:
    if k != 'Power':
        plt.scatter(df[k],df['Power'])
        plt.title(k)
        plt.yscale('log')  # Change the scale of the y-axis to logarithmic
        plt.xscale('log')  # Change the scale of the x-axis to logarithmic

        plt.show()
```

```python
for k in num:
    if k != 'Power':
        sns.distplot(df[k])
        plt.title(k)
        plt.show()
```

```python
for k in num:
    if k != 'Power':
        sns.boxplot(df[k])
        plt.title(k)
        plt.show()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:,-1].values
print(X)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

labelencoder = LabelEncoder()
X[:, 3] = labelencoder.fit_transform(X[:, 3])
onehotencoder = ColumnTransformer(
    transformers = [
        ("OneHot", #name
        OneHotEncoder(),# transformer class
        [3] #columns to be applied on
        )
    ],
    remainder = 'passthrough' # donot apply anything to the remaining columns
)
X = onehotencoder.fit_transform(X)
```

```python
X = X[:,1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.show()
```
